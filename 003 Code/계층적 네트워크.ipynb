```python
# 차 라벨 묶기 + 왼오 - 근데 멍청함
import cv2
import matplotlib.pyplot as plt
import random

# IOU 계산 함수
def calculate_iou(box1, box2):
    x1_min, y1_min, x1_max, y1_max = box1
    x2_min, y2_min, x2_max, y2_max = box2

    inter_x_min = max(x1_min, x2_min)
    inter_y_min = max(y1_min, y2_min)
    inter_x_max = min(x1_max, x2_max)
    inter_y_max = min(y1_max, y2_max)

    inter_area = max(0, inter_x_max - inter_x_min) * max(0, inter_y_max - inter_y_min)
    box1_area = (x1_max - x1_min) * (y1_max - y1_min)
    box2_area = (x2_max - x2_min) * (y2_max - y2_min)

    return inter_area / (box1_area + box2_area - inter_area) if (box1_area + box2_area - inter_area) > 0 else 0

# 포함 비율 계산 함수
def calculate_containment_ratio(outer_box, inner_box):
    x1_min, y1_min, x1_max, y1_max = outer_box
    x2_min, y2_min, x2_max, y2_max = inner_box

    inner_area = (x2_max - x2_min) * (y2_max - y2_min)
    outer_area = (x1_max - x1_min) * (y1_max - y1_min)

    if outer_area > 0:
        containment_ratio = inner_area / outer_area
        return containment_ratio >= 0.96
    return False

# 바운딩 박스 그룹화 함수
def group_boxes(boxes, labels, iou_min=0.02, iou_max=0.08):
    groups = []
    used = set()

    for i, box in enumerate(boxes):
        if i in used:
            continue
        group = [box]
        group_labels = [labels[i]]
        used.add(i)

        for j, other_box in enumerate(boxes):
            if j != i and j not in used:
                iou = calculate_iou(box, other_box)
                if iou_min <= iou <= iou_max:
                    group.append(other_box)
                    group_labels.append(labels[j])
                    used.add(j)

                if labels[i] == 0 and labels[j] == 1:  
                    if calculate_containment_ratio(box, other_box):
                        group.append(other_box)
                        group_labels.append(labels[j])
                        used.add(j)

        group.sort(key=lambda x: (x[2] - x[0]) * (x[3] - x[1]), reverse=True)
        groups.append((group, group_labels))

    return groups

# 라벨 업데이트 함수
def update_labels(grouped_boxes):
    updated_boxes = []

    for group, labels in grouped_boxes:
        if len(group) == 3:  # 그룹의 박스 수가 3개일 경우
            updated_labels = [labels[0], f"{labels[0]}-R", f"{labels[0]}-L"]
            updated_boxes.append((group, updated_labels))
        else:
            updated_boxes.append((group, labels))  # 변경 없음

    return updated_boxes


# 바운딩 박스 시각화 및 좌표 출력 함수
def visualize_labeled_boxes(image, grouped_boxes):
    img = image.copy()
    bounding_box_info = {}

    for idx, (group, labels) in enumerate(grouped_boxes):
        car_label = f"car{idx + 1}"
        bounding_box_info[car_label] = []

        for j, box in enumerate(group):
            x_min, y_min, x_max, y_max = map(int, box)

            # 색상 지정
            if j == 0:  # 첫 번째 차
                color = (0, 0, 255)  # 빨간색
            elif '-R' in str(labels[j]):  # R인 경우
                color = (255, 0, 0)  # 파란색
            elif '-L' in str(labels[j]):  # L인 경우
                color = (0, 255, 255)  # 노란색
            else:
                color = (0, 255, 0)  # 기본 초록색

            cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, 2)
            box_label = str(labels[j])  # 라벨을 문자열로 변환
            cv2.putText(img, box_label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
            bounding_box_info[car_label].append((box_label, box))

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # 바운딩 박스 좌표 출력
    for car, boxes in bounding_box_info.items():
        if len(boxes) == 3:  # 3개인 경우
            print(f"{car}: [{car}-1: {boxes[0][1]}, {car}-R: {boxes[1][1]}, {car}-L: {boxes[2][1]}]")
        else:  # 3개 미만인 경우
            print(f"{car}: [{', '.join([f'{car}-{j + 1}: {box[1]}' for j, box in enumerate(boxes)])}]")


# 바운딩 박스 데이터 추출 및 시각화
image_count = 0
results_count = len(results)

random_results = random.sample(results, min(5, results_count))

for result in random_results:
    if result.orig_img is not None and result.boxes is not None:
        img = result.orig_img.copy()
        boxes = result.boxes.xyxy.numpy()  
        labels = result.boxes.cls.numpy()  

        if len(boxes) == 0:
            print(f"No bounding boxes detected in image: {result.path}")
            continue

        grouped_boxes = group_boxes(boxes, labels)
        grouped_boxes = update_labels(grouped_boxes)  # 라벨 업데이트
        visualize_labeled_boxes(img, grouped_boxes)

        image_count += 1
```

```python
# 박스 겹칠때 깜빡이 그룹화 못하는거 조정
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np

# IOU 계산 함수
def calculate_iou(box1, box2):
    x1_min, y1_min, x1_max, y1_max = box1
    x2_min, y2_min, x2_max, y2_max = box2

    inter_x_min = max(x1_min, x2_min)
    inter_y_min = max(y1_min, y2_min)
    inter_x_max = min(x1_max, x2_max)
    inter_y_max = min(y1_max, y2_max)

    inter_area = max(0, inter_x_max - inter_x_min) * max(0, inter_y_max - inter_y_min)
    box1_area = (x1_max - x1_min) * (y1_max - y1_min)
    box2_area = (x2_max - x2_min) * (y2_max - y2_min)

    return inter_area / (box1_area + box2_area - inter_area) if (box1_area + box2_area - inter_area) > 0 else 0

# 바운딩 박스 그룹화 함수
def group_boxes(boxes, labels, confidences, iou_min=0.02, iou_max=0.08):
    groups = []
    used = set()

    for i, box in enumerate(boxes):
        if i in used or labels[i] != 0:  # 현재 박스가 차(라벨 0)가 아니면 스킵
            continue
        group = [(box, confidences[i])]  # 박스와 컨피던스를 함께 저장
        group_labels = [labels[i]]
        used.add(i)

        # 깜빡이 찾기
        blinkers = []
        for j, other_box in enumerate(boxes):
            if j != i and j not in used and labels[j] == 1:  # 깜빡이 라벨인 경우
                iou = calculate_iou(box, other_box)
                if iou_min <= iou <= iou_max:
                    blinkers.append((other_box, confidences[j]))
                    used.add(j)

        # 깜빡이 두 개까지만 추가
        blinkers = blinkers[:2]
        for blinker_box, blinker_conf in blinkers:
            group.append((blinker_box, blinker_conf))
            group_labels.append(labels[i])

        groups.append((group, group_labels))  # 그룹에 추가

    return groups

# 라벨 업데이트 함수
def update_labels(grouped_boxes):
    updated_boxes = []

    for idx, (group, labels) in enumerate(grouped_boxes):  # idx 사용
        car_label = f"car{idx + 1}"
        
        # 박스를 x_min을 기준으로 정렬하여 가장 왼쪽이 L, 오른쪽이 R이 되도록 함
        sorted_group = sorted(group, key=lambda item: item[0][0])  # item[0][0]은 x_min 좌표
        
        if len(sorted_group) == 3:  # 그룹의 박스 수가 3개일 경우
            updated_labels = [f"{car_label}-1", f"{car_label}-L", f"{car_label}-R"]
            updated_boxes.append((sorted_group, updated_labels))
        else:
            updated_labels = [f"{car_label}-{j + 1}" for j in range(len(group))]
            updated_boxes.append((sorted_group, updated_labels))

    return updated_boxes

# 바운딩 박스 시각화 및 좌표, 컨피던스 출력 함수
def visualize_labeled_boxes(image, grouped_boxes):
    img = image.copy()
    bounding_box_info = {}

    for idx, (group, labels) in enumerate(grouped_boxes):
        car_label = f"car{idx + 1}"
        bounding_box_info[car_label] = []

        for j, (box, confidence) in enumerate(group):
            x_min, y_min, x_max, y_max = map(int, box)

            # 색상 지정
            if j == 0:  # 첫 번째 차
                color = (0, 0, 255)  # 빨간색
            elif '-R' in str(labels[j]):  # R인 경우
                color = (255, 0, 0)  # 파란색
            elif '-L' in str(labels[j]):  # L인 경우
                color = (0, 255, 255)  # 노란색
            else:
                color = (0, 255, 0)  # 기본 초록색

            cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, 2)
            box_label = f"{labels[j]}: {confidence:.2f}"
            cv2.putText(img, box_label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 2)
            bounding_box_info[car_label].append((labels[j], box, confidence))

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # 바운딩 박스 좌표 및 컨피던스 점수 출력
    for car, boxes in bounding_box_info.items():
        # 좌표만 출력
        coords = ', '.join([f"{label}: {coords}" for label, coords, _ in boxes])
        print(f"{car}: [{coords}]")
        # 컨피던스 점수만 출력
        confidences = ', '.join([f"{label}: [{confidence:.2f}]" for label, _, confidence in boxes])
        print(f"{car}: [{confidences}]")

# 바운딩 박스 데이터 추출 및 시각화
image_count = 0
results_count = len(results)

random_results = random.sample(results, min(5, results_count))

for idx, result in enumerate(random_results):
    if result.orig_img is not None and result.boxes is not None:
        img = result.orig_img.copy()
        boxes = result.boxes.xyxy.numpy()  
        labels = result.boxes.cls.numpy()  
        confidences = result.boxes.conf.numpy()  # 컨피던스 점수

        if len(boxes) == 0:
            print(f"No bounding boxes detected in image: {result.path}")
            continue

        grouped_boxes = group_boxes(boxes, labels, confidences)
        grouped_boxes = update_labels(grouped_boxes)  # 라벨 업데이트
        visualize_labeled_boxes(img, grouped_boxes)

        image_count += 1
```
